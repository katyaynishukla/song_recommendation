# Popular songs list

```python
class popularity_recommender_py():
    def __init__(self):
        self.train_data = None
        self.user_id = None
        self.item_id = None
        self.popularity_recommendations = None
        
    def create(self, train_data, user_id, item_id):
        self.train_data = train_data
        self.user_id = user_id
        self.item_id = item_id

        train_data_grouped = train_data.groupby([self.item_id]).agg({self.user_id: 'count'}).reset_index()
        train_data_grouped.rename(columns = {'user_id': 'score'},inplace=True)
        train_data_sort = train_data_grouped.sort_values(['score', self.item_id], ascending = [0,1])
        train_data_sort['Rank'] = train_data_sort['score'].rank(ascending=0, method='first')
        self.popularity_recommendations = train_data_sort.head(50)

    def recommend(self, user_id):    
        user_recommendations = self.popularity_recommendations
        user_recommendations['user_id'] = user_id
        cols = user_recommendations.columns.tolist()
        cols = cols[-1:] + cols[:-1]
        user_recommendations = user_recommendations[cols]
        
        return user_recommendations
    
```

```python
pr = popularity_recommender_py() # pr is popularity recommender
```

```python
pr.create(song_df, 'user_id', 'song')
```

```python
pr = pr.recommend(song_df['user_id'][5])
pr
```

```python
popular_rating = song_df.merge(pr,on='song')
popular_rating = popular_rating.drop(['user_id_x', 'song_id','user_id_y','listen_count'], axis = 1)
popular_rating = popular_rating.drop_duplicates()
popular_rating = popular_rating.sort_values(by=['Rank'])
popular_rating = popular_rating.astype({'Rank':'int'})
popular_rating
```

```python
import pickle
pickle.dump(popular_rating,open('popular.pkl','wb'))
```

# Songs based on collaborative filtering

```python
x = song_df.groupby('user_id').count()['listen_count'] > 100
user_gave_rating = x[x].index
x
```

```python
filtered_rating = song_df[song_df['user_id'].isin(user_gave_rating)]
filtered_rating
```

```python
y = filtered_rating.groupby('title').count()['listen_count'] >= 50
famous_songs = y[y].index
famous_songs
```

```python
final_ratings = filtered_rating[filtered_rating['title'].isin(famous_songs)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='title',columns='user_id',values='listen_count')
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(song_name):
    index = np.where(pt.index==song_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = song_df_2[song_df_2['title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['artist_name'].values))
        
        data.append(item)
    
    return data
```

```python
recommend("Sehr kosmisch")
```
